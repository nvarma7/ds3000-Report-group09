## 2. METHOD

### 2.1. Data Acquisition

Our dataset is a [Kaggle](https://www.kaggle.com/tomigelo/spotify-audio-features) dataset. The dataset contains 13 different audio features of a song calculated and made available via the official Spotify Web API. Each song (row) has values for artist name, track name, track id and the audio features themselves. In total, the dataset has 130,663 rows and 17 columns. The data was scraped from Spotify on April 2019. 

*Audio Features*: 

1.   **Time signature**: An estimated overall time signature of a track. The time signature (meter) is a notational convention to specify how many beats are in each bar (or measure).
2.   **Energy**: A measure from 0.0 to 1.0 and represents a perceptual measure of intensity and activity. Typically, energetic tracks feel fast, loud, and noisy. For example, death metal has high energy, while a Bach prelude scores low on the scale. Perceptual features contributing to this attribute include dynamic range, perceived loudness, timbre, onset rate, and general entropy.
3. **Danceability**: Describes how suitable a track is for dancing based on a combination of musical elements including tempo, rhythm stability, beat strength, and overall regularity. A value of 0.0 is least danceable and 1.0 is most danceable. 
4. **Tempo**: The overall estimated tempo of a track in beats per minute (BPM). In musical terminology, tempo is the speed or pace of a given piece and derives directly from the average beat duration. 
5. **Valence**: A measure from 0.0 to 1.0 describing the musical positiveness conveyed by a track. Tracks with high valence sound more positive (e.g. happy, cheerful, euphoric), while tracks with low valence sound more negative (e.g. sad, depressed, angry).
6. **Liveness**: Detects the presence of an audience in the recording. Higher liveness values represent an increased probability that the track was performed live. A value above 0.8 provides strong likelihood that the track is live.
7. **Song Duration**: Duration of the track in milliseconds. 
8. **Acousticsness**: A confidence measure from 0.0 to 1.0 of whether the track is acoustic. 1.0 represents high confidence the track is acoustic.
9. **Instrumentalness**: Predicts whether a track contains no vocals. “Ooh” and “aah” sounds are treated as instrumental in this context. Rap or spoken word tracks are clearly “vocal”. The closer the instrumentalness value is to 1.0, the greater likelihood the track contains no vocal content. 
10. **Loudness**: The overall loudness of a track in decibels (dB). Loudness values are averaged across the entire track and are useful for comparing relative loudness of tracks. 
11. **Speechiness**: Speechiness detects the presence of spoken words in a track. The more exclusively speech-like the recording (e.g. talk show, audio book, poetry), the closer to 1.0 the attribute value. Values above 0.66 describe tracks that are probably made entirely of spoken words.
12. **Key**: The estimated overall key of the track. Integers map to pitches using standard Pitch Class notation . E.g. 0 = C, 1 = C♯/D♭, 2 = D, and so on. If no key was detected, the value is -1.
13. **Mode**: Mode indicates the modality (major or minor) of a track, the type of scale from which its melodic content is derived. Major is represented by 1 and minor is 0.
14. **Song Popularity**: Based on total number of plays compared to other tracks as well as how recent those plays are (calculated by Spotify).

####Multimodel Testing for Classification Algorithms 

```python
from sklearn.svm import LinearSVC
from sklearn.naive_bayes import GaussianNB
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier

estimators = {
    'k-Nearest Neighbor': KNeighborsClassifier(), # should match the method names you import 
    'Gaussian Naive Bayes': GaussianNB(),
    'Decision Tree': DecisionTreeClassifier()}
```

```python
from sklearn.model_selection import KFold
from sklearn.model_selection import cross_val_score

for estimator_name, estimator_object in estimators.items():
    kfold = KFold(n_splits =10, random_state =3000, shuffle = True)
    scores = cross_val_score(estimator=estimator_object, X=X_train,y=y_train,cv = kfold)
    estimator_object.fit(X=X_train, y=y_train)
    class_report = classification_report(y_true=y_train, y_pred=estimator_object.predict(X=X_train))
    
    print(estimator_name + ": \n\t" + f'mean accuracy={scores.mean():.2%}, ' + f'standard deviation={scores.std():.2%}'  + "\n\t"f'class report=\n{class_report}'  +"\n" )
```

```python
from scipy import stats

df_drop = data2.drop(["artist_name"], axis = 1)
df_drop = data2.drop(["track_id"], axis = 1)
df_drop = data2.drop(["track_name"], axis = 1)
df_col = df_drop.columns
df_array = df_drop.values
df_norm = normalize(data1)
normalized_df = pd.DataFrame(df_norm, columns = df_col)

def get_onewayANOVA(df, IV1, IV2):
    
    df_1 = 1
    df_2 = len(IV1)-1
    
    
    f_results = stats.f_oneway(df[IV1], df[IV2])
    #Formatting of F-Test
    f_stat = format(f_results[0],".02f")
    #Formatting of p-values from onewayANOVA
    p_val = format(f_results[1], ".03f")
    # Comducts F Statistics and P Values
    print("F(%s,%s) = %s, p = %s" % (df_1, df_2, f_stat, p_val))
    
    
    descriptives = df[IV1].agg(["mean", "sem"])
    print(IV1, "mean:", round(descriptives[0],100), "\nMSE:", round(descriptives[1],4))
    descriptives = df[IV2].agg(["mean", "sem"])
    print(IV1, "mean:", round(descriptives[0],100), "\nMSE:", round(descriptives[1],4))
```
